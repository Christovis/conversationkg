## Povenance
---

store to be able to recall where (e-mail & excerpt) a fact or entity was extracted from <br>
-> introduce two relations: EvidencedBy and MentionedBy, both between a fact or an entity and an e-mail (identified by hash value) <br>
-> make EvidencedBy and MentionedBy ternary relations, in which the third argument is the excerpt from the e-mail that the fact was drawn from 



### EvidencedBy

expresses facts and entities that occur meta to the conversation itself

| Fact/Entity |                    Origin |
|-------------|--------------------------:|
| Convo       | [$mail_1$, ..., $mail_n$] |
| Person      |                    $mail$ |
| Org         |                    Person |
| ...         |                       ... |


### MentionedBy

expresses fact and entities that occur inside the conversation and need not be factual otuside of it


| Fact/Entity  | Origin |
|--------------|-------:|
| Link/Address | $mail$ |
| Person       | $mail$ |
| Org          | $mail$ |
| Code snippet | $mail$ |
| ...          |    ... |

```python
import os
import pickle
from tqdm import tqdm

import re
import html

# from collections import Counter
# import matplotlib.pyplot as plt

from class_declarations import Email, Conversation, Ledger
```

```python
ledger = Ledger()
```

```python
mails = []
for f in tqdm(os.listdir("files")[:10000]):
    with open("files/"+f) as handle:
        mails.append(list(map(html.unescape, handle.readlines())))
```

```python
email_objs = [Email(m, ledger) for m in tqdm(mails)]
fltrd_emails = [m for m in email_objs if m.has_header]

with open("emails_provenance.pkl", "wb") as handle:
    pickle.dump(fltrd_emails, handle)

ledger.pickle()
```

```python
load = True
if load:
    with open("emails_provenance.pkl", "rb") as handle:
        fltrd_emails = pickle.load(handle)
    ledger = Ledger.from_pickle()
mails_srtd = sorted(fltrd_emails, key=lambda m: m.sent)
convos = Conversation.conversations_from_sorted_emails(mails_srtd, ledger)
```
